### Q3. Create a feature to differentiate 'Short', 'Medium', and 'Long' movies

```python
df1[df1['duration'] < 89]
```

```python
for i in df1.index:
    if df1.loc[i,'duration'] < 89:
        df1.loc[i,'interval'] = "Short"
    else:
        if df1.loc[i,'duration'] > 117:
            df1.loc[i,'interval'] = "Long"
        else:
            df1.loc[i,'interval'] = "Medium"
```

### Q5. Count of telecast in TV and movie show category

```python
df1['cast'].unique()
```

```python
cast_movies = df1['cast']
list1 = [x.split(", ") for x in cast_movies]
len(list1)
```

```python
cast_movies_all = []
for i in range(0,len(list1)):
    for item in list1[i]:
        if item not in cast_movies_all:
            cast_movies_all.append(item)
```

```python
cast_movies_all
len(cast_movies_all)
```

```python
df2 = df[df['type'] == 'TV Show']
cast_shows = df2['cast']
len(cast_shows)
list2 = [x.split(", ") for x in cast_shows]
len(list2)
```
